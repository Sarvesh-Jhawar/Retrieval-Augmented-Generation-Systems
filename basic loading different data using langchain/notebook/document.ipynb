### Data Ingestion

```python
# Document Structure
# ------------------
# We are importing the Document class from LangChain, which is the object type

# used to represent one piece of text (content + metadata) in the RAG pipeline.from langchain_core.documents import Document
# Think of Document as a simple container for text and its attributes.
```

```python
# Create a sample Document object
# --------------------------------
# Here we create one Document in memory with text and metadata.
# page_content: actual text that will be embedded/searched.
# metadata: extra context to track where this text came from.
from langchain_core.documents import Document

doc = Document(
    page_content="this is the main text content I am using to create RAG",
    metadata={
        "source": "example.txt",  # source file name

        "pages": 1,               # number of pages in sourcedoc

        "author": "Sarvesh Jhawar", # who created it (sample data)# Display the Document object to verify it looks correct

        "date_created": "2026-03-20" # creation date

    })
```

```python
# Create a directory for sample text files
# ----------------------------------------
# os.makedirs creates the folder path if it does not exist.

# exist_ok=True means it won't crash if the folder is already there.print("✅ Created directory ../data/text_files or it already existed")

import os
os.makedirs("../data/text_files", exist_ok=True)
```

```python
# Write sample text files for the demo
# -----------------------------------
# We create two text files with basic content, so later we can load and index them.

sample_texts = {
    "../data/text_files/python_intro.txt": """Python Programming Introduction

Python is a high-level, interpreted programming language known for its simplicity and readability.
Created by Guido van Rossum and first released in 1991, Python has become one of the most popular
programming languages in the world.

Key Features:
- Easy to learn and use
- Extensive standard library
- Cross-platform compatibility
- Strong community support

Python is widely used in web development, data science, artificial intelligence, and automation.""",

    "../data/text_files/machine_learning.txt": """Machine Learning Basics

Machine learning is a subset of artificial intelligence that enables systems to learn and improve
from experience without being explicitly programmed. It focuses on developing computer programs
that can access data and use it to learn for themselves.

Types of Machine Learning:
1. Supervised Learning: Learning with labeled data
2. Unsupervised Learning: Finding patterns in unlabeled data
3. Reinforcement Learning: Learning through rewards and penalties

Applications include image recognition, speech processing, and recommendation systems."""
}

# Save each string to its file path
for filepath, content in sample_texts.items():
    with open(filepath, 'w', encoding="utf-8") as f:
        f.write(content)

print("✅ Sample text files created!")
```

```python
# Load a single text file using TextLoader
# ---------------------------------------
# TextLoader reads the raw text from a local file and returns a list of Document objects.
# This is the first stage before embeddings/search.

from langchain_community.document_loaders import TextLoader

# Load file content and store as in-memory Document(s)

loader = TextLoader("../data/text_files/python_intro.txt", encoding="utf-8")

document = loader.load()# Show the loaded document to confirm everything is correct
print(document)
```

```python
# Directory Loader (load all text files in one go)
# -------------------------------------------------
# DirectoryLoader scans a folder for matching files, then uses TextLoader on each.
# This is useful when you have many documents and want to load them together.

from langchain_community.document_loaders import DirectoryLoader

# load all .txt files from the sample path
dir_loader = DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt",            # file name pattern (all .txt)
    loader_cls=TextLoader,       # loader class to create each Document
    loader_kwargs={'encoding': 'utf-8'},
    show_progress=False
)



documents = dir_loader.load()# Display the list of loaded Document objects
documents
```

```python
# Load PDF files via DirectoryLoader + PyMuPDFLoader
# --------------------------------------------------
# This reads PDF files from a folder and converts pages into Document objects.
# PDF content is often split by page, and metadata includes file path.

from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader

# load all .pdf files from the pdf folder
dir_loader = DirectoryLoader(
    "../data/pdf",
    glob="**/*.pdf",          # pattern for pdf files
    loader_cls=PyMuPDFLoader,   # loader class for reading PDF content
    show_progress=False

)

# Show loaded PDF documents

pdf_documents = dir_loader.load()
pdf_documents
```
